### **1**.**Defanging an IP address**

Given a valid (IPv4) IP address, return a defanged version of that IP address.

A defanged IP address replaces every period "." with "[.]".



```python
class Solution:
    def defangIPaddr(self, address: str) -> str:
        return address.replace('.','[.]')

#usage
new_addr=Solution()
print(new_addr.defangIPaddr("1.1.1.1"))
print(new_addr.defangIPaddr("255.100.50.0"))
```

### **2. Find numbers with even number of digits**
Given an array nums of integers, return how many of them contain an even number of digits.

```python
from typing import List
class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        #nums=[55,678,2,6,3476]
        n=len(nums)
        i=0
        count=0
        if 1<=n<=500:
            for i,idx in enumerate(nums):
                if 1<=nums[i]<=10**5:
                    dig_count=len(str(nums[i]))
                    if dig_count%2==0:
                      count+=1
                    i+=1
        return count


sol=Solution()
sol.findNumbers([5,4,5,2,0,44])
```

### **3. Number of good pairs**

Given an array of integers nums, return the number of good pairs.

A pair (i, j) is called good if nums[i] == nums[j] and i < j.

```python
from typing import List
class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        count=0
        j=1
        if 1 <= len(nums) <= 100:
          for i in range(len(nums)-1):
            if 1 <= nums[i] <= 100:
              for j in range(i+1,len(nums)):
                if nums[i]==nums[j] and i<j:
                    count+=1
                j+=1
            i+=1
        return count

sol=Solution()
sol.numIdenticalPairs([1,3,2,1,1,2])
```

### **4.How many numbers are smaller than the current number**

Given the array nums, for each nums[i] find out how many numbers in the array are smaller than it. That is, for each nums[i] you have to count the number of valid j's such that j != i and nums[j] < nums[i].

Return the answer in an array.

```python
import numpy as np
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        result=[0]*len(nums)

        sorted_lst=sorted(nums)
        if 2 <= len(nums) <= 500:
          for i in range(len(nums)):
            if 0 <= nums[i] <= 100:
              for j in range(len(nums)):
                if nums[i]==sorted_lst[j]:
                    result[i]=j
                    break
                j+=1
          i+=1
        return np.array(result)

sol=Solution()
sol.smallerNumbersThanCurrent([2,1,1,3,4])
```

### **5. Subtract the product and sum of the digits of an integer**

Given an integer number n, return the difference between the product of its digits and the sum of its digits.


```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        num=n
        digits=[]
        sum_num=0
        product_num=1
        while 1 <= num <= 10**5:
            digit = num % 10  # Get last digit
            digits.append(digit)
            num = num // 10  # Remove last digit
        #print(digits[::-1])  # Reverse to get correct order

        for i in range(len(digits)):
            sum_num=sum_num+digits[i]
            product_num=product_num*digits[i]
            #print(sum_num)
            #print(product_num)
            #print(product_num-sum_num)
        return(product_num-sum_num)

sol=Solution()
sol.subtractProductAndSum(345)
```

### **6. XOR operation in an array**

Define an array nums where nums[i] = start + 2 * i (0-indexed) and n == nums.length.

Return the bitwise XOR of all elements of nums.

```python
import numpy as np
class Solution:
    def xorOperation(self, n: int, start: int) -> int:

      if 1 <= n <= 1000 and 0 <= start <= 1000:
        # Initialize nums as a NumPy array of the correct size
        nums = np.zeros(n, dtype=int)
        for i in range(n):
          nums[i]=start+2*i
# After the loop, the full array is constructed
        print(f"Final nums array: {nums}")
        result_xor = 0
        for x in nums:
          result_xor ^= x
        print(f"Bitwise XOR result: {result_xor}")
sol=Solution()
sol.xorOperation(6,0)
```
