### **1. Maximum product of two elements in an array**

Given the array of integers nums, you will choose two different indices i and j of that array. Return the maximum value of (nums[i]-1)*(nums[j]-1).

```python
from typing import List
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
    # Largest two: max1 >= max2
        max1 = max2 = 0
        if 2 <= len(nums) <= 500:
            for num in nums:
                if 2 <= len(nums) <= 500 and 1<=num<=10**3:
  # Update largest two
                    if num >= max1:
                        max2 = max1
                        max1 = num
                    elif num > max2:
                        max2 = num
        return((max1 - 1) * (max2 - 1) )
sol=Solution()
sol.maxProduct([1,5,4,5])
```

### **2. Count number of teams**

There are n soldiers standing in a line. Each soldier is assigned a unique rating value.

You have to form a team of 3 soldiers amongst them under the following rules:

Choose 3 soldiers with index (i, j, k) with rating (rating[i], rating[j], rating[k]).
A team is valid if: (rating[i] < rating[j] < rating[k]) or (rating[i] > rating[j] > rating[k]) where (0 <= i < j < k < n).
Return the number of teams you can form given the conditions. (soldiers can be part of multiple teams).



```python
class Solution:
    def numTeams(self, rating):
        n = len(rating)
        count = 0

        # For each soldier as the middle (j)
        for j in range(1, n - 1):
          if 3<=n<=1000 and 1<=rating[j]<=10**5:
            left_less = left_greater = 0
            right_less = right_greater = 0

            # Count soldiers to the left of j
            for i in range(j):
                if rating[i] < rating[j]:
                    left_less += 1
                elif rating[i] > rating[j]:
                    left_greater += 1

            # Count soldiers to the right of j
            for k in range(j + 1, n):
                if rating[k] < rating[j]:
                    right_less += 1
                elif rating[k] > rating[j]:
                    right_greater += 1

            # Add valid teams with j as middle soldier
            count += left_less * right_greater  # increasing teams
            count += left_greater * right_less  # decreasing teams

        return count
sol=Solution()
sol.numTeams([2,1,3])

```

### **3. Number of students doing homework at a given time**

Given two integer arrays startTime and endTime and given an integer queryTime.

The ith student started doing their homework at the time startTime[i] and finished it at time endTime[i].

Return the number of students doing their homework at time queryTime. More formally, return the number of students where queryTime lays in the interval [startTime[i], endTime[i]] inclusive.

```python
from typing import List
class Solution:
    def busyStudent(self, startTime: List[int], endTime: List[int], queryTime: int) -> int:
        count = 0
        if len(startTime)==len(endTime) and 1 <= queryTime <= 1000:
          n = len(startTime)  # All arrays have same length

          for i in range(n):
            if 1 <= n <= 100 and 1 <= startTime[i] <= endTime[i] <= 100:
              if startTime[i] <= queryTime <= endTime[i]:
                count += 1

          return count
sol=Solution()
sol.busyStudent([2,3,4],[4,2,3],6)
```

```python
def busyStudent(startTime, endTime, queryTime):
    return sum(1 for s, e in zip(startTime, endTime) if s <= queryTime <= e)
busyStudent([2,3,4],[4,6,6],6)
```

### **4. Number of steps to reduce a number to zero**

Given an integer num, return the number of steps to reduce it to zero.

In one step, if the current number is even, you have to divide it by 2, otherwise, you have to subtract 1 from it.

```python
class Solution:
    def numberOfSteps(self, num: int) -> int:
      if 0<=num<=10**6:
        count=0
        while num>0:
            if num%2==0:
                num=num/2
                count+=1
            else:
                num=num-1
                count+=1
        return count
sol=Solution()
sol.numberOfSteps(123)
```

### **5. Counting bits**

Given an integer n, return an array ans of length n + 1 such that for each i (0 <= i <= n), ans[i] is the number of 1's in the binary representation of i.

```python
class Solution:
    def countBits(self, n: int) -> List[int]:
      if 0<=n<=10**5:
        ans = []
        for i in range(n + 1):
            # Count '1's in binary representation of i
            ans.append(bin(i).count('1'))
        return ans
sol=Solution()
sol.countBits(10)
```
